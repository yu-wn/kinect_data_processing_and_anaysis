```python
#フレームを指定して姿勢をプロット
import pandas as pd
import plotly.graph_objects as go
import plotly.io as pio
import math 

file_path = "C:\\Users\\miyazakipc\\Desktop\\2024年9月10日　実験\\Kinect\\1回目\\pos_20240910182038.csv"

N = 150
df = pd.read_csv(file_path, header=None)
data = df.to_numpy()
# 鼻、耳、目のデータは表示していない
Body_x = data[N, 2:81:3]
Body_y = data[N, 3:82:3]
Body_z = data[N, 4:83:3]

list1 = Body_x.tolist()
list2 = Body_y.tolist()
list3 = Body_z.tolist()

fig = go.Figure()

# 点のプロット
fig.add_trace(go.Scatter3d(
    x=list1,
    y=list2,
    z=list3,
    mode='markers',
    marker=dict(size=2, color='black'),
    name='Points'
))

# 線のプロット
lines = [
    (0, 1), (1, 2), (2, 3), (4, 5), (5, 6), (6, 7), (7, 8), (8, 9),
    (11, 12), (12, 13), (13, 14), (14, 15), (15, 16), (18, 19), (19, 20),
    (20, 21), (22, 23), (23, 24), (2, 4), (7, 10), (2, 11), (14, 17),
    (0, 18), (0, 22), (24, 25), (3, 26)
]

for start, end in lines:
    fig.add_trace(go.Scatter3d(
        x=[list1[start], list1[end]],
        y=[list2[start], list2[end]],
        z=[list3[start], list3[end]],
        mode='lines',
        line=dict(color='gray', width=3),
        name='Lines'
    ))

# 原点のプロット
fig.add_trace(go.Scatter3d(
    x=[0],
    y=[0],
    z=[0],
    mode='markers',
    marker=dict(size=5, color='blue'),
    name='Origin'
))
# x軸方向の単位ベクトル
fig.add_trace(go.Scatter3d(
    x=[0,1],
    y=[0,0],
    z=[0,0],
    mode='lines',
    line=dict(color='blue', width=3),
    name='X-Axis'
))

# y軸方向の単位ベクトル
fig.add_trace(go.Scatter3d(
    x=[0,0],
    y=[0,1],
    z=[0,0],
    mode='lines',
    line=dict(color='blue', width=3),
    name='Y-Axis'
))

# z軸方向の単位ベクトル
fig.add_trace(go.Scatter3d(
    x=[0,0],
    y=[0,0],
    z=[0,1],
    mode='lines',
    line=dict(color='blue', width=3),
    name='Z-Axis'
))
# レイアウトを設定(正面)
fig.update_layout(
    scene=dict(
        xaxis_title='X Axis',
        yaxis_title='Y Axis',
        zaxis_title='Z Axis',
        camera=dict(
            up=dict(x=0.21, y=-1.1, z=2),  # カメラの上方向を設定
            center=dict(x=0, y=0, z=0),    # カメラの中心位置を設定
            eye=dict(x=-0.2, y=-0.8, z=-2) # カメラの視点を設定
        )
    )
)

fig.show()
```

```python
#時系列データ可視化
#コパイロット作成版
import pandas as pd
import plotly.graph_objects as go

# CSVファイルのパス
file_path = "C:\\Users\\miyazakipc\\Desktop\\2024年9月10日　実験\\Kinect\\1回目\\pos_20240910182038.csv"

# データの読み込み
df = pd.read_csv(file_path, header=None)
data = df.to_numpy()

# フレーム数の取得
num_frames = data.shape[0]

# 各フレームのデータを取得
frames = [{
    'x': data[N, 2:81:3],
    'y': data[N, 3:82:3],
    'z': data[N, 4:83:3]
} for N in range(num_frames)]

# 点と線を結ぶための接続リスト
lines = [
    (0, 1), (1, 2), (2, 3), (4, 5), (5, 6), (6, 7), (7, 8), (8, 9),
    (11, 12), (12, 13), (13, 14), (14, 15), (15, 16), (18, 19), (19, 20),
    (20, 21), (22, 23), (23, 24), (2, 4), (7, 10), (2, 11), (14, 17),
    (0, 18), (0, 22), (24, 25), (3, 26)
]

# 3Dプロットの作成
fig = go.Figure()

# フレームを生成
animation_frames = []
for i, frame in enumerate(frames):
    scatter = go.Scatter3d(
        x=frame['x'],
        y=frame['y'],
        z=frame['z'],
        mode='markers',
        marker=dict(size=3, color='black'),
        name=f'Frame {i}'
    )
    lines_data = []
    for start, end in lines:
        lines_data.append(go.Scatter3d(
            x=[frame['x'][start], frame['x'][end]],
            y=[frame['y'][start], frame['y'][end]],
            z=[frame['z'][start], frame['z'][end]],
            mode='lines',
            line=dict(color='gray', width=3),
            name='Lines'
        ))
    animation_frames.append(go.Frame(data=[scatter] + lines_data, name=f'Frame {i}'))

# 初期フレームのデータを設定
fig.add_trace(animation_frames[0].data[0])
for trace in animation_frames[0].data[1:]:
    fig.add_trace(trace)

# アニメーションの設定
fig.update_layout(
    updatemenus=[{
        'buttons': [
            {
                'args': [None, {'frame': {'duration': 50, 'redraw': True}, 'mode': 'immediate'}],
                'label': 'Play',
                'method': 'animate'
            },
            {
                'args': [[None], {'frame': {'duration': 0, 'redraw': True}, 'mode': 'immediate'}],
                'label': 'Pause',
                'method': 'animate'
            }
        ],
        'type': 'buttons',
        'showactive': False,
        'direction': 'left',
        'x': 0.1,
        'xanchor': 'right',
        'y': 0,
        'yanchor': 'top'
    }],
    scene=dict(
        xaxis_title='X Axis',
        yaxis_title='Y Axis',
        zaxis_title='Z Axis',
        camera=dict(
            up=dict(x=0.21, y=-1.1, z=2),
            center=dict(x=0, y=0, z=0),
            eye=dict(x=-0.2, y=-0.8, z=-2)
        )
    ),
    sliders=[{
        'active': 0,
        'yanchor': 'top',
        'xanchor': 'left',
        'currentvalue': {
            'font': {'size': 20},
            'prefix': 'Frame:',
            'visible': True,
            'xanchor': 'right'
        },
        'steps': [{
            'args': [
                [f'Frame {k}'],
                {'frame': {'duration': 0, 'redraw': True}, 'mode': 'immediate'}
            ],
            'label': str(k),
            'method': 'animate'
        } for k in range(num_frames)]
    }]
)

# フレームを更新
fig.update(frames=animation_frames)

fig.show()
```

```python
import pandas as pd
import numpy as np

# データの読み込み
file_path = "C:\\Users\\miyazakipc\\Desktop\\2024年9月10日　実験\\Kinect\\1回目\\pos_20240910182038.csv"
df = pd.read_csv(file_path, header=None)
data = df.to_numpy()

# 各フレームのデータを取得
num_frames = data.shape[0]
frames = [{
    'x': data[N, 2:81:3],
    'y': data[N, 3:82:3],
    'z': data[N, 4:83:3]
} for N in range(num_frames)]

# 異常値の閾値設定
threshold = 0.9  # 変化量の閾値
cleaned_frames = [frames[0]]  # 初期フレームを追加

# フレーム間の変化量を計算して、異常値を除去
for i in range(1, num_frames):
    diff_x = frames[i]['x'] - frames[i-1]['x']
    diff_y = frames[i]['y'] - frames[i-1]['y']
    diff_z = frames[i]['z'] - frames[i-1]['z']
    
    if np.all(np.abs(diff_x) <= threshold) and np.all(np.abs(diff_y) <= threshold) and np.all(np.abs(diff_z) <= threshold):
        cleaned_frames.append(frames[i])

# cleaned_framesに異常値が除去されたデータが格納されます。
import plotly.graph_objects as go

# 3Dプロットの作成
fig = go.Figure()

# アニメーションのフレームを生成
animation_frames = []
for i, frame in enumerate(cleaned_frames):
    scatter = go.Scatter3d(
        x=frame['x'],
        y=frame['y'],
        z=frame['z'],
        mode='markers',
        marker=dict(size=3, color='black'),
        name=f'Frame {i}'
    )
    lines_data = []
    for start, end in lines:
        lines_data.append(go.Scatter3d(
            x=[frame['x'][start], frame['x'][end]],
            y=[frame['y'][start], frame['y'][end]],
            z=[frame['z'][start], frame['z'][end]],
            mode='lines',
            line=dict(color='gray', width=3),
            name='Lines'
        ))
    animation_frames.append(go.Frame(data=[scatter] + lines_data, name=f'Frame {i}'))

# 初期フレームのデータを設定
fig.add_trace(animation_frames[0].data[0])
for trace in animation_frames[0].data[1:]:
    fig.add_trace(trace)

# アニメーションの設定
fig.update_layout(
    updatemenus=[{
        'buttons': [
            {
                'args': [None, {'frame': {'duration': 50, 'redraw': True}, 'mode': 'immediate'}],
                'label': 'Play',
                'method': 'animate'
            },
            {
                'args': [[None], {'frame': {'duration': 0, 'redraw': True}, 'mode': 'immediate'}],
                'label': 'Pause',
                'method': 'animate'
            }
        ],
        'type': 'buttons',
        'showactive': False,
        'direction': 'left',
        'x': 0.1,
        'xanchor': 'right',
        'y': 0,
        'yanchor': 'top'
    }],
    scene=dict(
        xaxis_title='X Axis',
        yaxis_title='Y Axis',
        zaxis_title='Z Axis',
        camera=dict(
            up=dict(x=0.21, y=-1.1, z=2),
            center=dict(x=0, y=0, z=0),
            eye=dict(x=-0.2, y=-0.8, z=-2)
        )
    ),
    sliders=[{
        'active': 0,
        'yanchor': 'top',
        'xanchor': 'left',
        'currentvalue': {
            'font': {'size': 20},
            'prefix': 'Frame:',
            'visible': True,
            'xanchor': 'right'
        },
        'steps': [{
            'args': [
                [f'Frame {k}'],
                {'frame': {'duration': 0, 'redraw': True}, 'mode': 'immediate'}
            ],
            'label': str(k),
            'method': 'animate'
        } for k in range(len(cleaned_frames))]
    }]
)

# フレームを更新
fig.update(frames=animation_frames)

fig.show()

```
