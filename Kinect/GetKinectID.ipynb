```python
import pandas as pd
from pathlib import Path
#最も多いIDを取得し、保存する
# 親フォルダのパスを指定
parent_folder = Path("C:\\Users\\miyazakipc\\Desktop\\酒井さんデータ\\高齢者実験\\kinect\\kourei_data\\kinect_20221121")

# 出力先フォルダのベースディレクトリを指定
output_base_dir = Path("C:\\Users\\miyazakipc\\Desktop\\酒井さんデータ\\高齢者実験\\kinect\\kourei_dataID\\kinect_20221121")
output_base_dir.mkdir(parents=True, exist_ok=True)

# 再帰的にフォルダ内のすべてのCSVファイルを検索
csv_files = list(parent_folder.rglob("*.csv"))

# CSVファイルを1つずつ処理
for file_path in csv_files:
    try:
        # CSVファイルを読み込む
        data = pd.read_csv(file_path, header=None)
        data.columns = [f"Col_{i}" for i in range(data.shape[1])]  # 仮の列名を付与
        
        # B列（2列目）をIDとして扱う
        data.rename(columns={"Col_1": "ID"}, inplace=True)

        # IDごとにデータを分割
        id_groups = data.groupby("ID")
        
        # データ数の最も多いIDを見つける
        largest_group_id = id_groups.size().idxmax()
        largest_group = id_groups.get_group(largest_group_id)

        # 出力先ディレクトリをファイルごとに設定
        relative_path = file_path.relative_to(parent_folder)
        output_dir = output_base_dir / relative_path.parent
        output_dir.mkdir(parents=True, exist_ok=True)
        
        # 最も多いIDのデータのみを保存
        output_file = output_dir / f"{relative_path.stem}_ID_{largest_group_id}.csv"
        largest_group.to_csv(output_file, index=False, encoding="utf-8-sig")
        
        print(f"処理完了: {file_path}, 最も多いID: {largest_group_id}")
    except Exception as e:
        print(f"エラーが発生しました: {file_path} - {e}")

print(f"すべてのCSVファイルの処理が完了しました。出力ディレクトリ: {output_base_dir}")
```
