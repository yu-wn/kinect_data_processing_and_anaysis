```python
import pandas as pd
import os

# Kinectの抽出：指定インデックスから指定フレームを抽出
# フォルダのパス
input_folder_path = "C:\\Users\\miyazakipc\\Desktop\\酒井さんデータ\\若年者\\kinect\\EachMotion\\Subject6\\normal"
output_folder_base_path = "C:\\Users\\miyazakipc\\Desktop\\酒井さんデータ\\data"

# フォルダが存在しない場合に作成
if not os.path.exists(output_folder_base_path):
    os.makedirs(output_folder_base_path)

# ここに入力
start_row_numbers = [243, 215, 203, 210, 193, 206, 218, 170, 194]

row_numbers = [
    num if num - 60 < 0 else num - 60
    for num in start_row_numbers
]

frames_to_extract = 120  # 抽出するフレーム数

# フォルダ内の .csv ファイルのみを取得
files = sorted([file for file in os.listdir(input_folder_path) if file.endswith(".csv")])

# デバッグ用出力
print(f"取得したファイル: {files}")
print(f"ファイル数: {len(files)}")
print(f"row_numbers: {row_numbers}")
print(f"row_numbers の数: {len(row_numbers)}")

# ファイル数と row_numbers の一致確認
if len(files) != len(row_numbers):
    print(f"エラー: ファイル数 ({len(files)}) と row_numbers の数 ({len(row_numbers)}) が一致していません。")
    exit()

# 各ファイルごとに処理
for i in range(len(row_numbers)):  # row_numbers の長さでループ
    file = files[i]
    input_csv_path = os.path.join(input_folder_path, file)
    
    # 入力フォルダの階層構造を取得
    relative_path = os.path.relpath(input_csv_path, input_folder_path)
    
    # 出力フォルダのパスを作成
    output_csv_path = os.path.join(output_folder_base_path, relative_path)
    output_csv_dir = os.path.dirname(output_csv_path)
    
    # 出力フォルダが存在しない場合に作成
    if not os.path.exists(output_csv_dir):
        os.makedirs(output_csv_dir)
    
    row_number = row_numbers[i]

    # データの読み込み
    df = pd.read_csv(input_csv_path, header=None)

    # 抽出したいフレーム数分だけ行を選択
    extract_start_row = max(0, row_number - frames_to_extract + 1)
    df_extract = df.iloc[extract_start_row:row_number + 1]

    # データの保存
    df_extract.to_csv(output_csv_path, index=False, header=None)

    print(f"{input_csv_path}のデータが {output_csv_path} に保存されました。")

print("すべてのファイルの処理が完了しました。")
```
